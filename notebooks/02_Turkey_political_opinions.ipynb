**HDP**: The Peoples' Democratic Party, or Democratic Party of the Peoples, is a pro-minority political party in Turkey. Generally left-wing, the party places a strong emphasis on participatory and radical democracy, feminism, LGBT+ rights, minority rights, youth rights and egalitarianism.
